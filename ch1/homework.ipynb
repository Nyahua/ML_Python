```python
import numpy as np
import pandas as pd

data1 = pd.read_table("ReportCard1.txt",header=0,delim_whitespace=True,index_col=None)
data2 = pd.read_table("ReportCard2.txt",header=0,delim_whitespace=True,index_col=None)
data = pd.merge(data1, data2, on='xh', how = 'outer')
data.info()
```

通过read_table函数分别从"ReportCard1.txt"、"ReportCard2.txt"读入学生信息,将两部分合并为一个表格。

```python
data['mean'] = np.array(data.iloc[:,2:9].mean(axis=1, skipna=True))
data['mean'] = data['mean'].apply(lambda x:'{:.2f}'.format(x))
data['total score'] = np.array(data.iloc[:,2:9].sum(axis = 1,skipna = True))
data.info()
```

利用iloc函数确定求和及求均值的列（即表示各学科得分的列），分别利用mean、sum函数求出成绩的均值、综合，函数中坐标系参数设置为对一行求均值、和，跳过NA,使用apply函数lambda形式保留两位小数

```python
data.sort_values(by='total score',ascending=False)[:]
```
